```python
# Generate random data representing daily temperatures over one year
data = np.random.normal(loc=20.0, scale=5.0, size=365)

# Create a histogram of the data with specified bins, color, and transparency
plt.hist(data, bins=20, color='blue', alpha=0.7)

# Set the title of the plot
plt.title('Histogram of Population per country ')

# Set the label for the x-axis
plt.xlabel('Country name')

# Set the label for the y-axis
plt.ylabel('Population')

# Display grid lines on the plot
plt.grid(True)

# Show the plot
plt.show()
```

```python
Country = ['Zimbabwe', 'Afghanistan', 'Algeria', 'Italy']
Population= [792207, 296032, 736350, 356244]
# Creating pie chart
plt.figure(figsize=(7, 7))
plt.pie(Population, labels=Country, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Pupulation share')  
plt.show()
```

```python
plt.scatter(Population, Country)
plt.title('Scatter Plot of Population vs. Country')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
# Data
countries = ['Indonesia', 'Brazil', 'United States', ' China', 'Pakistan', 'India', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico']
population = [1444216107, 1393409038, 332915073, 276361783, 225199937, 214280581, 211400708, 166303498, 145912025, 130262216]

# Sort data
population, countries = zip(*sorted(zip(population, countries), reverse=True))

# Funnel plot
plt.figure(figsize=(10, 6))
plt.fill_betweenx(range(len(countries)), 0, population, color="skyblue", alpha=0.4)
plt.plot(population, range(len(countries)), marker='o', color="skyblue", alpha=0.6)

# Labels
plt.yticks(range(len(countries)), countries)
plt.title('Funnel Plot of Population by Country')
plt.xlabel('Population')
plt.ylabel('Country')
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Sample data
Population = [100, 200, 300, 400, 500]

# Creating a boxplot
plt.figure(figsize=(6, 5))
plt.boxplot(Population)
plt.title('Boxplot of Population')
plt.ylabel('Population')

# Show the plot
plt.show()
```

```python
Population= [792207, 296032, 736350, 356244]

# Convert the list of years to numpy array
Population_array = np.array(Population)

# Calculate summary statistics
mean = np.mean(Population_array)
median = np.median(Population_array)
std_dev = np.std(Population_array)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```
